```python
import numpy as np
import matplotlib.pyplot as plt
from opensolar.api import get_kWh_production_dummy
from opensolar.utils import create_date_series
import datetime

# Create a sample dataset
start_date = datetime.date(2023, 1, 1)
end_delta = datetime.timedelta(weeks=52)
dates = create_date_series(start_date, end_delta)

data = np.array([i + 1 for i in range(366)])
random_delta = np.random.normal(0, 10)
kWhs = [get_kWh_production_dummy(0, 0, date, random_delta) for date in dates]


# Plot the scaled data
plt.plot(data, kWhs)
plt.xlabel('Index')
plt.ylabel('Scaled Value (0 to pi)')
plt.title('Scaled Data')
plt.show()
```
