```python
from opensolar.utils import load_google_cloud_key
from opensolar.api import get_address_infos
from opensolar.api import  get_google_maps_image
from opensolar.utils import show_image

api_key = load_google_cloud_key()
address = "Rainweg 6, Neuenstadt"
address_infos = get_address_infos(address, api_key=api_key)

if len(address_infos) == 1:
    address_info = address_infos[0]
    formatted_address=address_info["formatted_address"]
    latitude = address_info["latitude"]
    longitude = address_info["longitude"]
    print(formatted_address)
    print(longitude, latitude)
    image = get_google_maps_image(latitude, longitude, api_key=api_key, zoom=20)
    show_image(image)
else:
   print("There are multiple options please pick!") 
```
